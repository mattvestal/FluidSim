```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datu = np.loadtxt('u.dat')
datv = np.loadtxt('v.dat')
datT = np.loadtxt('T.dat')
top = np.loadtxt('topo.dat')
```

```python
x = np.linspace(0,101,101)
y = np.linspace(0,51,51)
X,Y=np.meshgrid(x,y)
```

```python
#51x101 grids, 192 outputs (every 15 minutes)
ind = np.arange(0,51) + 151*51
```

```python
ind
```

```python
Uuse = []
Vuse = []
Tuse = []
```

```python
for i in ind:
    u = datu[i]
    v = datv[i]
    t = datT[i]
    Uuse.append(u)
    Vuse.append(v)
    Tuse.append(t)
Uuse = np.array(Uuse)
Vuse = np.array(Vuse)
Tuse = np.array(Tuse)
```

```python
levels = np.arange(0,0.6.0.05)
ig,ax = plt.subplots(figsize=(10,10))
ax.quiver(X,Y,Uuse,Vuse)
ax.contour(X,Y,Tuse,0.05)
plt.show()
```
